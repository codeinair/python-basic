```python
import pandas as pd
filePath = 'D:\\Personal\\Study\\Bits Pilani\\Git\\python-basic-1\\Regression\\air-quality\\AirQuality.xlsx'
airQualityData = pd.read_excel(filePath,index_col=0)
```

```python
# replace -200 values with mean value in column "NOx(GT)"
temp = airQualityData[airQualityData["NOx(GT)"] != -200]["NOx(GT)"]
airQualityData["NOx(GT)"] = airQualityData["NOx(GT)"].replace(-200,temp.mean())

# replace -200 values with mean value in column "CO(GT)"
temp = airQualityData[airQualityData["CO(GT)"] != -200]["CO(GT)"]
airQualityData["CO(GT)"] = airQualityData["CO(GT)"].replace(-200,temp.mean())

del(temp)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math

print("Number of total input records: " + str(len(airQualityData)))

#split the data into training and testing
#split_factor = 0.70
trainFactor = int(0.07 * len(airQualityData))

#setting the matrices
X_train = airQualityData.values[:trainFactor, 1:] 
y_train = airQualityData.values[:trainFactor, 0:1].reshape((-1, 1)) 
X_test = airQualityData.values[trainFactor:, 1:] 
y_test = airQualityData.values[trainFactor:, 0:1].reshape((-1, 1))

#Building Simple linear regression model
reg = LinearRegression() 
reg.fit(X_train, y_train)
print("Coefficient: " + str(reg.coef_) + " Intercept: " + str(reg.intercept_))
print("As we see that the Coefficient value is not zero, there exists a relationship between CO(GT) and NOx(GT).")
```
